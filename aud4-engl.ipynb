## z-test

**Null hypothesis:** Socio-economic class of people did not affect the survival rate.

**Alternative hypothesis:** People's socio-economic class affected their survival rate.

```python
from statsmodels.stats.weightstats import ztest as ztest
```

```python
FirstClass = df.query('Pclass == 1')
ThirdClass = df.query('Pclass == 3')

FirstClass_sample = FirstClass.sample(n=100, random_state=1) #random_state pamti koji sample je izabrani
ThirdClass_sample = ThirdClass.sample(n=100, random_state=1)
```

```python
df.groupby('Pclass')['Survived'].mean() #show the mean of Survived for each value in Pclass
```

```python
FirstClass_sample['Survived'].hist()
```

```python
Survived_normal = np.array([np.mean(df[df["Pclass"]==1].sample(20)["Survived"].values) for i in range(100)])
FirstClass_sample['Survived_normal'] = Survived_normal

Survived_normal_t = np.array([np.mean(df[df["Pclass"]==3].sample(20)["Survived"].values) for i in range(100)])
ThirdClass_sample['Survived_normal'] = Survived_normal_t

#samples 20 passengers from the first class, calculates the survival rate for each sample, 
#and stores these survival rates in the "Survived_normal" column of the FirstClass_sample dataframe
# same for third class
```

```python
FirstClass_sample['Survived_normal'].hist() #histogram of the survival rates in first class
```

## Two Sample Z-Test 

```python
ztest(FirstClass_sample['Survived_normal'], ThirdClass_sample['Survived'], value=0) 
```

## One Sample Z-test

```python
ztest(FirstClass_sample['Survived_normal'], value=0)
```

```python
ztest(ThirdClass_sample['Survived_normal'], value=0)
```

## Decide which test to use

**Two basic categories:**

1. **Parametric:** assumes that the sample data come from a population that can be adequately modeled by a probability distribution that has a fixed set of parameters

2. **Non-parametric:** does not assume an explicit (finite parametric) mathematical form for the distribution when modeling data. However, it may make some assumptions about that distribution, such as continuity or symmetry.

```python
from scipy import stats
```

## Homogeneity

```python
stats.levene(FirstClass_sample['Survived_normal'],ThirdClass_sample['Survived_normal'])
```

## T-test

```python
stats.ttest_ind(FirstClass_sample['Survived_normal'],ThirdClass_sample['Survived_normal'])
```

## Analysis of Variance Test (ANOVA)

Tests whether the means of two or more independent samples are significantly different.

**Assumptions**

1. Observations in each sample are independent and identically distributed (iid).
2. Observations in each sample are normally distributed.
3. Observations in each sample have the same variance.

**Interpretation**

1. H0: the means of the samples are equal.
2. H1: one or more of the means of the samples are unequal.

```python
from scipy.stats import f_oneway
stat, p = f_oneway(FirstClass_sample['Survived_normal'],ThirdClass_sample['Survived_normal'])
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably the same distribution')
else:
	print('Probably different distributions')
```

## Shapiro-Wilk Test

**Tests whether a data sample has a Gaussian distribution.**

**Assumptions:**

Observations in each sample are independent and identically distributed (iid).

1. H0: the sample has a Gaussian distribution.
2. H1: the sample does not have a Gaussian distribution.

```python
from scipy.stats import shapiro
```

```python
stat, p = shapiro(FirstClass_sample['Survived_normal'])
```

```python
print('stat=%.3f, p=%.3f' % (stat, p))
```

```python
if p > 0.05:
	print('Probably Gaussian')
else:
	print('Probably not Gaussian')
```

## Mann-Whitney U Test

Tests whether the distributions of two independent samples are equal or not.

Assumptions

Observations in each sample are independent and identically distributed (iid).
Observations in each sample can be ranked.
Interpretation

H0: the distributions of both samples are equal.
H1: the distributions of both samples are not equal.

```python
from scipy.stats import mannwhitneyu
stat, p = mannwhitneyu(FirstClass_sample['Survived_normal'],ThirdClass_sample['Survived_normal'])
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably the same distribution')
else:
	print('Probably different distributions')
```

## Wilcoxon Signed-Rank Test

**Tests whether the distributions of two paired samples are equal or not.**

**Assumptions:**

1. Observations in each sample are independent and identically distributed (iid).
2. Observations in each sample can be ranked.
3. Observations across each sample are paired.

**Interpretation**

1. H0: the distributions of both samples are equal.
2. H1: the distributions of both samples are not equal.

```python
from scipy.stats import wilcoxon
stat, p = wilcoxon(FirstClass_sample['Survived_normal'],ThirdClass_sample['Survived_normal'])
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably the same distribution')
else:
	print('Probably different distributions')
```

## Kruskal-Wallis H Test

Tests whether the distributions of two or more independent samples are equal or not.

**Assumptions**

1. Observations in each sample are independent and identically distributed (iid).
2. Observations in each sample can be ranked.

**Interpretation**

1. H0: the distributions of all samples are equal.
2. H1: the distributions of one or more samples are not equal.

```python
from scipy.stats import kruskal
stat, p = kruskal(FirstClass_sample['Survived_normal'],ThirdClass_sample['Survived_normal'])
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably the same distribution')
else:
	print('Probably different distributions')
```
