```python
# For the features EthnicGroup, ParentMaritalStatus and WklyStudyHours replace NaN values with the most common value.

most_common_ethnic_group = X['EthnicGroup'].mode()[0]
X['EthnicGroup'].fillna(most_common_ethnic_group, inplace=True)

most_common_parent_marital_status = X['ParentMaritalStatus'].mode()[0]
X['ParentMaritalStatus'].fillna(most_common_parent_marital_status, inplace=True)

most_common_weekly_study_hours = X['WklyStudyHours'].mode()[0]
X['WklyStudyHours'].fillna(most_common_weekly_study_hours, inplace=True)
```

```python
# For the feature ParentEduc replace NaN value with a string no formal education.
X['ParentEduc'].fillna('no formal education', inplace=True)
```

```python
# For the feature TestPrep replace NaN value with a string none.
X['TestPrep'].fillna('none', inplace=True)
```
